Di bawah ini visual data

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data_frame['Grade'])
plt.title('Distribution Grade Table') 
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Survival Months', y='Status', data=data_frame)
plt.title('Scatter plot Survival Months vs Status')
```

```python
%matplotlib widget
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
plot_geeks = ax.scatter(data_frame['Age'], data_frame['Tumor Size'], data_frame['Survival Months'], color='red')
ax.set_xlabel('Age')
ax.set_ylabel('Tumor Size')
ax.set_zlabel('Survival Months')
plt.figure(figsize=(10, 6))
plt.title('3D Stage')
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

output_notebook()

TOOLTIPS = HoverTool(tooltips = [
    ('index', '$index'),
    ('(Survival Months, Status)', '(@Survival Months, @Status)'),
    ('Age', '@Age')
])

p = figure(title = 'Data Cancers', x_axis_label = 'Survival Months', y_axis_label = 'Status', width =700, height =700, tools=[TOOLTIPS])
p.scatter('Survival Months', 'Status', size=10, source=data_frame)
show(p)
```

```python
data = sns.load_dataset('iris')
sns.displot(data['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Sepal Length vs Sepal width')
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("pastel")

sns.boxplot(x='species', y='sepal_length', data=data)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
